# Regresion Lineal

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv("./datasets/netflix_titles-score.csv")
data
```

```python
lista1 = list(data['show_id'])
lista3 = []
```

```python
for n in lista1:
    lista3.append('n' + n)

```

```python
data.insert(loc=0,column='id',value=lista3)
```

```python
data
```

```python
data_isnull = data.fillna({'rating': 'G'}) # sin necesidad de hacer copia seria con el implace=true
```

```python
data1 = data_isnull
```

```python
data1
```

```python
data1['description'] = data1['description'].str.lower() # convertir los datos en minusculas
```

```python
# primero se comprueba que la columna sea un dato de tipo str y luego se usa la funcion split para separar en dos columnas y guardarlos en una variable para luego concatenar con el df
duration = data1["duration"].str.split(expand=True)
duration.columns = ['duration_int', 'duration_type']
data1 = pd.concat([data1, duration], axis=1)
data1
```

```python
type(data1['duration_int'][0])
```

```python
data1['duration_int']= data1['duration_int'].astype(int, copy=True, errors='raise')
```

```python
data1
```

```python
data1['duration_type'] = data1['duration_type'].replace({'Seasons':'Season'}) # reempkazamos las palabras Seasons por season
```

```python
data1 = data1.reindex(columns=['id','show_id','type','title','director','cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'duration_int', 'duration_type', 'listed_in', 'description', 'score'])
```

```python
data1 = data1.drop(['duration'], axis=1) # eliminamos la columna duration 
```

```python
data1 = data1.drop(['show_id'], axis=1)
```

```python
data1['date_added'] = pd.to_datetime(data1.date_added) # cambiar el formato de fecha
```

```python
data1.to_csv("./datasets/netflix_titles-score_Arreglado.csv", index=False)
```
